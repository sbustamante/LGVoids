```python
#==================================================================================================
#Importing all necessary libraries
#==================================================================================================
%pylab inline
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
# JSAnimation import available at https://github.com/jakevdp/JSAnimation
from JSAnimation import IPython_display
from matplotlib import animation
import numpy as np
import os
from matplotlib.colors import colorConverter
#My Cmaps
cmap2 = mpl.colors.LinearSegmentedColormap.from_list('my_cmap2',["white","blue"],2)
cmap2._init()
alphas = np.ones( cmap2.N+3)
alphas[0] = 0
cmap2._lut[:,-1] = alphas
my_cmap4 = plt.cm.get_cmap('gray', 4)

#Folder where is stored all simulation data
data_folder = '../../../CosmicData/BOLSHOI/'
#Folder where is stored all binary codes (C)
binary_folder = './'
#Parameter of the simulation
Lbox = 250.0    #[Mpc]
Nbox = 256

#Trheshold parameters for the cosmic web
Tl_th = 0.265
Vl_th = 0.175

#==================================================================================================
#Functions
#==================================================================================================
#Cutting Density field in X/Y/Z axe
def CutField( filename, X, res=32, Coor = 3, N=Nbox ):
    coord = [3,2,1]
    if res == 'plain':
        os.system( "./Field_Cut_Plain.out %s %d temp.tmp %d %d"%( filename, X, coord[Coor-1], N ) )
        datos = np.loadtxt( 'temp.tmp' )
        N = int(np.sqrt(len( datos )))
        datos = datos.reshape( (N, N) )
        os.system( "rm temp.tmp" )
      
    else:	
        os.system( "./Field_Cut%d.out %s %d temp.tmp %d"%( res, filename, X, coord[Coor-1] ) )
        datos = np.loadtxt( 'temp.tmp' )
        N = int(np.sqrt(len( datos )))
        datos = datos.reshape( (N, N) )
        os.system( "rm temp.tmp" )
    return datos

#Field to compute Environment of Cosmic web
def Scheme( eig1, eig2, eig3, Lamb ):
    N1 = len( eig1 )
    N2 = len( eig1[0] )
    sch = 0*eig1
    
    for i in xrange(N1):
        for j in xrange(N2):
            if eig1[i,j] > Lamb: sch[i,j] += 1
            if eig2[i,j] > Lamb: sch[i,j] += 1
            if eig3[i,j] > Lamb: sch[i,j] += 1
    return sch

#==================================================================================================
#Loading all data (catalogues of halos and voids)
#==================================================================================================
#BDM catalogue of halos
BDM_halos = np.loadtxt( "%s/C_GH_BDM.dat"%(data_folder) )
#FOF catalogue of halos
FOF_halos = np.loadtxt( "%s/C_GH_FOF.dat"%(data_folder) )

#BDM catalogue of isolated pairs
BDM_ip = np.loadtxt( "%s/C_IP_BDM.dat"%(data_folder) )
#FOF catalogue of isolated pairs
FOF_ip = np.loadtxt( "%s/C_IP_FOF.dat"%(data_folder) )

#BDM catalogue of reduced isolated pairs
BDM_ip = np.loadtxt( "%s/C_IP_BDM.dat"%(data_folder) )
#FOF catalogue of reduced isolated pairs
FOF_ip = np.loadtxt( "%s/C_IP_FOF.dat"%(data_folder) )

#Catalogue of distances of BDM halos to Tweb-FAG-21 voids
#    Twe-FAG-21 : Void finder based on a Watershed transform over the FA field of the Tweb scheme
#    with a 2nd-order median filtering applied and with boundary removal activated
dist_Tweb_BDM = np.loadtxt( "%s/Tweb/256/C_GH-voids_s1_BDM_FAG.dat"%(data_folder) )

#Filename of density field
density_f = "%s/Tweb/256/Delta_s1"%(data_folder)
#Filename of Tweb eigenvalues
eigenT_f = "%s/Tweb/256/Eigen_s1"%(data_folder)
#Filename of Vweb eigenvalues
eigenV_f = "%s/Vweb/256/Eigen_s1"%(data_folder)

#Folder of Tweb-FAG-21 voids
voidsT21_f = "%s/Tweb/256/voidsFAG/voids_21/"%(data_folder)
```

**Function for plotting a portion of the cosmic web**

```python
#plot_web
#This function plot certain cubic region of the cosmic web
def plot_web( c, ext, eigen, l_th ):
    
    X = []; Y = []; Z = []; C = []
    #Sweeping X coordinates
    for i in xrange( c[0]-ext, c[0]+ext+1 ):
        #Periodic boundaries
        ir = i
        if i<0: ir+=Nbox
        if i>Nbox: ir-=Nbox
        it = i-(c[0]-ext)
            
        #Calculating visual impresion of this X-slide
        datos = Scheme(CutField( eigen+"_1", ir, 16, 0 ), CutField( eigen+"_2", ir, 16, 0 ), CutField( eigen+"_3", ir, 16, 0 ), l_th )
        
        #Sweeping Y coordinates
        for j in xrange( c[1]-ext, c[1]+ext+1 ):
            #Periodic boundaries
            jr = j
            if j<0: jr+=Nbox
            if j>Nbox: jr-=Nbox
            jt = j-(c[1]-ext)
            
            #Sweeping Z coordinates
            for k in xrange( c[2]-ext, c[2]+ext+1 ):
                #Periodic boundaries
                kr = k
                if k<0: kr+=Nbox
                if k>Nbox: kr-=Nbox
                kt = k-(c[2]-ext)
                    
                #Storing visual impression
                X.append( it )
                Y.append( jt )
                Z.append( kt )
                C.append( datos[jr,kr] )
    return np.array(X), np.array(Y), np.array(Z), np.array(C)
```

```python
fig = plt.figure( figsize = (15,15) )
ax = fig.add_subplot(111, projection='3d')
X, Y, Z, C = plot_web( [50,50,50], 20, eigenT_f, Tl_th )
ax.scatter(X, Y, Z, c=C, marker='o', s = 20, linewidth=0.01, alpha=0.5 )
#P = 3
#ax.scatter(X[C==P], Y[C==P], Z[C==P], c=C[C==P], marker='o', s = 10, linewidth=0.01 )
ax.set_xlim( (0,41) )
ax.set_ylim( (0,41) )
ax.set_zlim( (0,41) )
ax.set_axis_off()
```

**Function for locating the place of some halo in the cosmic web**

```python
#halo_environment
def halo_environment( idh, ext ):
    
    #Loading cell coordinates of current halo
    c = (BDM_halos[idh,1:4]/Lbox*Nbox).astype(int)
    #Loading distance to the nearest void
    void_halo = dist_Tweb_BDM[ idh ]
    void_c = void_halo[2:5].astype(int)
    print "Halo cell", c, "\nVoid cell", void_halo[2:5].astype(int), "\nDistance", void_halo[0]
    
    #Name of eigenvalues
    eigen = eigenT_f
    l_th = Tl_th
    
    #Indexes
    ind = [0,1,2,0,1,2]
    #Axis
    axes = ["Z", "Y", "X"]
    
    XX = (BDM_halos[:,1]/Lbox*Nbox).astype(int)
    YY = (BDM_halos[:,2]/Lbox*Nbox).astype(int)
    ZZ = (BDM_halos[:,3]/Lbox*Nbox).astype(int)
    
    plt.figure( figsize = (23,7) )
    
    for coor in xrange(3):
        X = np.zeros( (2*ext+1,2*ext+1) )
        V = np.zeros( (2*ext+1,2*ext+1) )
        #Calculating visual impresion of this "X"-slide
        datos = Scheme(CutField( eigen+"_1", c[ind[coor]], 16, coor+1 ), CutField( eigen+"_2", c[ind[coor]], 16, coor+1 ), 
        CutField( eigen+"_3", c[ind[coor]], 16, coor+1 ), l_th )
        #Loading the nearest void
        void = CutField( "%s/void_index.dat"%(voidsT21_f), c[ind[coor]], "plain", coor+1 )
        void[void!=void_halo[1].astype(int)] = 0 
        
        plt.subplot( 1,3,coor+1 )
        #Visual impresion
        if coor == 0:
            #Sweeping X coordinates
            for j in xrange( c[2]-ext, c[2]+ext+1 ):
                #Periodic boundaries
                jr = j
                if j<0: jr+=Nbox
                if j>=Nbox: jr-=Nbox
                jt = j-(c[2]-ext)
                #Sweeping Y coordinates
                for k in xrange( c[1]-ext, c[1]+ext+1 ):
                    #Periodic boundaries
                    kr = k
                    if k<0: kr+=Nbox
                    if k>=Nbox: kr-=Nbox
                    kt = k-(c[1]-ext)
                    #Storing visual impression
                    X[jt,kt] = datos[jr,kr]
                    V[jt,kt] = void[jr,kr]
                
            plt.plot( [c[2],void_c[2]], [c[1],void_c[1]], '-', color=(0.0,1,0.0), linewidth = 3, zorder=10 )
            plt.plot( [c[2],], [c[1],], 'o', color='yellow', markersize = 8, zorder=5 )
            plt.plot( XX[ZZ==c[0]], YY[ZZ==c[0]], '.', markersize=4, color = 'red' )
            plt.imshow( -np.transpose(X)[::-1,::], extent=( c[2]-ext, c[2]+ext+1, c[1]-ext, c[1]+ext+1 ), 
            cmap=my_cmap4 )
            plt.imshow( np.transpose(V)[::-1,::], extent=( c[2]-ext, c[2]+ext+1, c[1]-ext, c[1]+ext+1 ), 
            cmap=cmap2 )
            
            
        if coor == 1:
            #Sweeping X coordinates
            for j in xrange( c[2]-ext, c[2]+ext+1 ):
                #Periodic boundaries
                jr = j
                if j<0: jr+=Nbox
                if j>=Nbox: jr-=Nbox
                jt = j-(c[2]-ext)
                #Sweeping Z coordinates
                for k in xrange( c[0]-ext, c[0]+ext+1 ):
                    #Periodic boundaries
                    kr = k
                    if k<0: kr+=Nbox
                    if k>=Nbox: kr-=Nbox
                    kt = k-(c[0]-ext)
                    #Storing visual impression
                    X[jt,kt] = datos[jr,kr]
                    V[jt,kt] = void[jr,kr]
                    
            plt.plot( [c[2],void_c[2]], [c[0],void_c[0]], '-', color=(0.0,1,0.0), linewidth = 3, zorder=10 )
            plt.plot( [c[2],], [c[0],], 'o', color='yellow', markersize = 8, zorder=5 )
            plt.plot( XX[YY==c[1]], ZZ[YY==c[1]], '.', markersize=4, color = 'red' )
            plt.imshow( -np.transpose(X)[::-1,::], extent=( c[2]-ext, c[2]+ext+1, c[0]-ext, c[0]+ext+1 ), 
            cmap=my_cmap4 )
            plt.imshow( np.transpose(V)[::-1,::], extent=( c[2]-ext, c[2]+ext+1, c[0]-ext, c[0]+ext+1 ), 
            cmap=cmap2 )
            
            
        if coor == 2:
            #Sweeping Y coordinates
            for j in xrange( c[1]-ext, c[1]+ext+1 ):
                #Periodic boundaries
                jr = j
                if j<0: jr+=Nbox
                if j>=Nbox: jr-=Nbox
                jt = j-(c[1]-ext)
                #Sweeping Z coordinates
                for k in xrange( c[0]-ext, c[0]+ext+1 ):
                    #Periodic boundaries
                    kr = k
                    if k<0: kr+=Nbox
                    if k>=Nbox: kr-=Nbox
                    kt = k-(c[0]-ext)
                    #Storing visual impression
                    X[jt,kt] = datos[jr,kr]
                    V[jt,kt] = void[jr,kr]
                    
            plt.plot( [c[1],void_c[1]], [c[0],void_c[0]], '-', color=(0.0,1,0.0), linewidth = 3, zorder=10 )
            plt.plot( [c[1],], [c[0],], 'o', color='yellow', markersize = 8, zorder=5 )
            plt.plot( YY[XX==c[2]], ZZ[XX==c[2]], '.', markersize=4, color = 'red' )
            plt.imshow( -np.transpose(X)[::-1,::], extent=( c[1]-ext, c[1]+ext+1, c[0]-ext, c[0]+ext+1 ), 
            cmap=my_cmap4 )
            plt.imshow( np.transpose(V)[::-1,::], extent=( c[1]-ext, c[1]+ext+1, c[0]-ext, c[0]+ext+1 ), 
            cmap=cmap2 )
            
        #Format
        plt.title( "%s slide [%s = %d]"%(axes[coor],axes[coor],c[ind[coor]]) )
        plt.xlabel( "SG %s"%(axes[ind[coor+1]]) )
        plt.ylabel( "SG %s"%(axes[ind[coor+2]]) )

    plt.subplot
    return V
```
