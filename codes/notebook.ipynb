```python
#==================================================================================================
#General functions
#==================================================================================================
#Cmap for Web
my_cmap4 = plt.cm.get_cmap('gray', 4)
#Customize 2map
def cmap2(color):
    cmap2b = mpl.colors.LinearSegmentedColormap.from_list('my_cmap2',["white",color],2)
    cmap2b._init()
    alphas = np.ones( cmap2b.N+3)
    alphas[0] = 0
    cmap2b._lut[:,-1] = alphas
    return cmap2b

#Classification Scheme
def Scheme( eig1, eig2, eig3, Lamb ):
    N1 = len( eig1 )
    try:
        N2 = len( eig1[0] )
        sch = 0*eig1
        for i in xrange(N1):
            for j in xrange(N2):
                if eig1[i,j] > Lamb: sch[i,j] += 1
                if eig2[i,j] > Lamb: sch[i,j] += 1
                if eig3[i,j] > Lamb: sch[i,j] += 1
        return sch
    except:
        sch = 0*eig1
        for i in xrange(N1):
            if eig1[i] > Lamb: sch[i] += 1
            if eig2[i] > Lamb: sch[i] += 1
            if eig3[i] > Lamb: sch[i] += 1
        return sch
        

#Cut Field
def CutFieldZ( filename, X, res=32, Coor = 3, N=256 ):
    '''
    Coor  1 -- X      2 -- Y      3 -- Z
    '''
    if res == 'plain':
	os.system( "./Field_Cut_Plain.out %s %d temp.tmp %d %d"%( filename, X, Coor, N ) )
	datos = np.loadtxt( 'temp.tmp' )
	N = int(np.sqrt(len( datos )))
	datos = datos.reshape( (N, N) )
	os.system( "rm temp.tmp" )
      
    else:	
	os.system( "./Field_Cut%d.out %s %d temp.tmp %d"%( res, filename, X, Coor ) )
	datos = np.loadtxt( 'temp.tmp' )
	N = int(np.sqrt(len( datos )))
	datos = datos.reshape( (N, N) )
	os.system( "rm temp.tmp" )
    return datos
```

## Isolated Pairs

Local Group candidates in Bolshoi are going to be selected from the general fof catalogue of dark matter halos following the next criteria:

- Each halo has a mass in the range $7\times 10^{11} M_{\odot}<M_{h}<7\times 10^{12} M_{\odot}$, consistent with the estimated values for MW and M31.
- Regarding each halo, there cannot be any other halo within the mass range $7\times 10^{11} M_{\odot}<M_{h}<7\times 10^{12} M_{\odot}$ closer than its companion.
- The relative radial velocity between both halos must be negative.
- The distance between the centre of mass of the halos must be less than $1.0\ Mpc$.
- There cannot be halos more massive than $7\times 10^{12} M_{\odot}$ within a radius of $3\ Mpc$ to every object centre.
- There cannot be halos more massive than $7\times 10^{13} M_{\odot}$ within a radius of $4\ Mpc$ to every object centre (Virgo-like systems).

For this selection we are going to use the code **[Pair-Finder](https://github.com/sbustamante/Pair-Finder)**. This code produces a catalogue of general pairs (i.e. pairs of halos that are the closest one to each other), and an isolated sample with all the previous criteria.

```python
#Loading produced Isolated-Pair catalogue (id1,M1,id2,M2,rrel,vrad,vtan)
isopairs = np.loadtxt( "%s/isolated_pairs_fof.dat"%(datafold), usecols=(1,2,4,5,7,8,9) )
#Number of halos
Nisop = len(isopairs)
#Calculating tangential velocity
isopairs[:,6] = np.sqrt( isopairs[:,6]**2 - (isopairs[:,5]-100*isopairs[:,4])**2 )

#Properties of halo1 of each isolated pair
halos1 = halos[ (isopairs[:,0]-1).astype(int) ]
#Properties of halo2 of each isolated pair
halos2 = halos[ (isopairs[:,2]-1).astype(int) ]
```

## LG candidates with compatible orbit

```python
#Energy-Angular momentum of a pair system
def Energy_AngularM( M1, x1, y1, z1, vx1, vy1, vz1, M2, x2, y2, z2, vx2, vy2, vz2):
 
    #km to MPC
    KM2MPC = 3.24077929e-20
    #kg to MSUN
    KG2MSUN = 5.05e-31
    #Cavendish
    G = 6.6742e-11*(1e-3*KM2MPC)**3/KG2MSUN
 
    E = np.zeros( len(M1) )
    L = np.zeros( (len(M1),3) )
    for i in xrange( len(M1) ):    
        m1 = M1[i]/hubble
        r1 = np.array( (x1[i],y1[i],z1[i]) )/hubble
        v1 = np.array( (vx1[i],vy1[i],vz1[i]) )
        
        m2 = M2[i]/hubble
        r2 = np.array( (x2[i],y2[i],z2[i]) )/hubble
        v2 = np.array( (vx2[i],vy2[i],vz2[i]) )
        r12 = r2-r1
        
        hubble_flow = 100*r12
        v12 = v2-v1 + hubble_flow
        
        L[i] = np.cross( r12, v12 )
        E[i] = 0.5*sum( v12*v12 )*KM2MPC**2 - G*(m1 + m2)/sqrt(sum(r12*r12))
    
    return E, L
```

```python
#Histogram of Radial vs Tangential velocities
#Limits
Elim = [-10e13,0]
Llim = [0,30e13]
Elim = [-40,0]
Llim = [0,250]

#no labels
nullfmt = NullFormatter()

#definitions for the axes
left, width = 0.1, 0.65
bottom_v, height = 0.1, 0.65
bottom_h = left_h = left+width+0.02
rect_hist2D = [left, bottom_v, width, height]
rect_histx = [left, bottom_h, 1.*width, 0.2]
rect_histy = [left_h, bottom_v, 0.2, height]
#start with a rectangular Figure
plt.figure(1, figsize=(8,8))
axHist2D = plt.axes(rect_hist2D)
axHistx = plt.axes(rect_histx)
axHisty = plt.axes(rect_histy)
#no labels
axHistx.xaxis.set_major_formatter(nullfmt)
axHisty.yaxis.set_major_formatter(nullfmt)

#Calculating energy and L
Eip, Lip = Energy_AngularM( 
halos1[:,7], halos1[:,1], halos1[:,2], halos1[:,3], halos1[:,4], halos1[:,5], halos1[:,6],\
halos2[:,7], halos2[:,1], halos2[:,2], halos2[:,3], halos2[:,4], halos2[:,5], halos2[:,6])
Eip = Eip/1e-36
Lip = np.sqrt(Lip[:,0]**2 + Lip[:,1]**2 + Lip[:,2]**2)

Elg, Llg = Energy_AngularM(
lghalos1[:,7], lghalos1[:,1], lghalos1[:,2], lghalos1[:,3], lghalos1[:,4], lghalos1[:,5], lghalos1[:,6],\
lghalos2[:,7], lghalos2[:,1], lghalos2[:,2], lghalos2[:,3], lghalos2[:,4], lghalos2[:,5], lghalos2[:,6])
Elg = Elg/1e-36
Llg = np.sqrt(Llg[:,0]**2 + Llg[:,1]**2 + Llg[:,2]**2)

#2D histogram
scatter2d = axHist2D.plot( Eip, Lip, "o" )
scatter2d = axHist2D.plot( Elg, Llg, "o", color="red" )

#Histogram X
histx = np.histogram( Eip, bins=np.sqrt(Nisop), range=Elim )
axHistx.bar( histx[1][:-1], histx[0], width = abs(Elim[1]-Elim[0])/np.sqrt(Nisop), linewidth=2.0, color="gray" )
#Histogram Y
histy = np.histogram( Lip, bins=np.sqrt(Nisop), range=Llim )
axHisty.barh( histy[1][:-1], histy[0], height = (Llim[1]-Llim[0])/np.sqrt(Nisop), linewidth=2.0, color="gray" )

#Format
axHist2D.set_xlabel( "e$_{tot}$ [$10^{-36}$Mpc$^2/$s$^2$]", fontsize=14 )
axHist2D.set_ylabel( "l$_{orb}$ [Mpc km/s]", fontsize=14 )
axHist2D.set_xlim( Elim )
axHist2D.set_ylim( Llim )
axHist2D.grid()

axHistx.grid( color='black', linestyle='--', linewidth=1., alpha=0.3 )
axHisty.grid( color='black', linestyle='--', linewidth=1., alpha=0.3 )
```

```python
#Plotting environment
def VisualWeb( axe = 3, coor = 0, axes = None ):
    #Loading axis
    if axes == None:
        fig = plt.figure()
        axes = fig.add_subplot(1,1,1)
        
    #Loading Fields
    eig1 = CutFieldZ( eig_filename+"_1", coor, 16, Coor = axe )
    eig2 = CutFieldZ( eig_filename+"_2", coor, 16, Coor = axe )
    eig3 = CutFieldZ( eig_filename+"_3", coor, 16, Coor = axe )

    #Plotting
    axes.imshow( -Scheme( eig1, eig2, eig3, l_thr )[:,::-1].T, extent = [0,Lbox,0,Lbox], vmin=-3, vmax=0,
    cmap = my_cmap4, interpolation = 'none' )
    axes.set_xlabel( "%s [$h^{-1}\ Mpc$]"%(xlabels[axe-1]), fontsize=14 )
    axes.set_ylabel( "%s [$h^{-1}\ Mpc$]"%(ylabels[axe-1]), fontsize=14 )
    axes.set_title("%s"%(web), fontsize=14 )
    axes.legend( )
    return axes
```

```python
#Effective radius function
def r_eff(X):
    return ((10**np.log10(X)*(0.9765625)**3)/( 4*np.pi/3. ))**(1/3.)
```

## Identifying the nearest voids of each LG system

```python
#halo_environment
def halo_environment( catalog, idvoid, halos_eigs, idh, dx=None, Nneigh=1 ):
    #Loading discrete indexes of the halo
    ih = int(np.floor(catalog[idh,1]*res/Lbox))
    jh = int(np.floor(catalog[idh,2]*res/Lbox))
    kh = int(np.floor(catalog[idh,3]*res/Lbox))
    inh = [ih,jh,kh]
    #Loading web filename
    eig_filename = '%s/Eigen_s1'%(bolshoidata)

    #Plotting projections in each plane
    fig = plt.figure( figsize=(15,6) )
    
    #Plane XY============================================================================================
    for index in xrange(3):
        subplot = fig.add_subplot(1,3,index+1)
        #Indexes of the other coordinates
        i1 = xindex[index]
        j1 = yindex[index]
        
        #Plotting current halo
        subplot.plot( [catalog[idh,i1],], [catalog[idh,j1],], "o", color="red", ms=6 )
        
        #Plotting catalogue of halos
        filt = (catalog[:,index+1]>catalog[idh,index+1]-1)&(catalog[:,index+1]<catalog[idh,index+1]+1)
        color = -Scheme( halos_eigs[filt,0], halos_eigs[filt,1], halos_eigs[filt,2], l_thr )
        subplot.scatter( [catalog[filt,i1],], [catalog[filt,j1],], c=color, vmin=-3, vmax=0,\
        cmap = my_cmap4, s=14, lw=0.2 )
        
        #Plotting environment of LG
        VisualWeb( axe = index+1, coor = inh[index], axes=subplot )
        
        #Catalogue of voids
        voids = CutFieldZ( "%s/void_index_Tweb.dat"%(datafold), inh[index], 'plain', Coor=index+1, N=res )
        #Plotting the closer neighbours
        for ingb in xrange(Nneigh):
            #Filtering no belonging voids
            void = voids==idvoid[idh,ingb]
            #Plotting voids
            subplot.imshow( np.transpose(void[::,::-1]), extent = [0,Lbox,0,Lbox], interpolation = 'none',\
            cmap=cmap2(colors[ingb]) )

        #Format
        subplot.set_title( "Halo %d (%s = %1.2f)"%\
        (catalog[idh,0], zlabels[index],catalog[idh,index+1]), fontsize=14 )
        
        if dx != None:
            subplot.set_xlim( (catalog[idh,i1]-dx,catalog[idh,i1]+dx) )
            subplot.set_ylim( (catalog[idh,j1]-dx,catalog[idh,j1]+dx) )
        else:
            subplot.set_xlim( (0,Lbox) )
            subplot.set_ylim( (0,Lbox) )
        
    
    return None
```
